```python
import cv2
import numpy as np
import face_recognition
import urllib
from google.colab.patches import cv2_imshow
```

```python
url1='https://raw.githubusercontent.com/dscoool/opencv/refs/heads/main/elon_musk.jpeg' #일론머스크 이미지
url2='https://raw.githubusercontent.com/dscoool/opencv/refs/heads/main/face.jpeg' #보통사람(정재민) 이미지 2
url3='https://raw.githubusercontent.com/dscoool/opencv/refs/heads/main/face2.jpeg' #일론머스크 이미지 3


img1 = urllib.request.urlopen(url1)
imgElon = face_recognition.load_image_file(img1) #인코딩에 사용할 이미지 가져오기
imgElon = cv2.cvtColor(imgElon,cv2.COLOR_BGR2RGB)  #RGB로 변환

img2 = urllib.request.urlopen(url2)
imgTest2 = face_recognition.load_image_file(img2)  #Test2 이미지 가져오기
imgTest2 = cv2.cvtColor(imgTest2,cv2.COLOR_BGR2RGB) #Tset2 이미지 RGB로 변환

img3 = urllib.request.urlopen(url3)
imgTest3 = face_recognition.load_image_file(img3)  #Test3 이미지 가져오기
imgTest3 = cv2.cvtColor(imgTest3,cv2.COLOR_BGR2RGB) #Tset3 이미지 RGB로 변환

```

```python
#얼굴의 위치 같게 만들기
faceLoc = face_recognition.face_locations(imgElon)[0]
encodeElon = face_recognition.face_encodings(imgElon)[0]   #감지할 얼굴 인코딩, 첫번째 요소만 가져오기
cv2.rectangle(imgElon,(faceLoc[3],faceLoc[0]), (faceLoc[1],faceLoc[2]),(255,0,255),2) #우리가 얼굴을 감지한 위치를 확인하기 위해 사각형을 이미지에 그림

```

#### imgtest2 - 정재민 이미지를 *사용*

```python
faceLocTest = face_recognition.face_locations(imgTest2)[0]
encodeTest = face_recognition.face_encodings(imgTest2)[0]   #Test이미지에 대한 첫번째 요소만 가져오기
cv2.rectangle(imgTest2,(faceLocTest[3],faceLocTest[0]), (faceLocTest[1],faceLocTest[2]),(255,0,255),2) #Test 이미지에 사각형 이미지
```

#### imgtest3 - 일론머스크3 이미지를 *사용*

```python
faceLocTest = face_recognition.face_locations(imgTest3)[0]
encodeTest = face_recognition.face_encodings(imgTest3)[0]   #Test이미지에 대한 첫번째 요소만 가져오기
cv2.rectangle(imgTest3,(faceLocTest[3],faceLocTest[0]), (faceLocTest[1],faceLocTest[2]),(255,0,255),2) #Test 이미지에 사각형 이미지
```

```python
results = face_recognition.compare_faces([encodeElon],encodeTest) #인코딩 이미지와 Test 이미지 간의 비교하기
faceDis = face_recognition.face_distance([encodeElon],encodeTest) #이미지 유사성 알기, (==> 얼굴 간의 오차 느낌인거 같음)

print(results, faceDis) #두 개의 이미지가 서로 같으면 True, 다른 이미지일 경우 False를 출력한다


#이미지에 대한 결과랑 유사성을 Test 이미지에 명시해주기, round()는 유사성을 소수점 둘째짜리로 반올림 한다는 뜻
cv2.putText(imgTest3, f'{results} {round(faceDis[0], 2)}', (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 0, 255), 2)

cv2_imshow(imgElon) #인코딩 이미지 불러오기
cv2_imshow(imgTest3) #Test 이미지 불러오기
cv2.waitKey(0)
```

```python
def blur_face(img):
  face_locations = face_recognition.face_locations(img)
  for (top, right, bottom, left) in face_locations:
    face_image = img[top:bottom, left:right]
    face_image = cv2.GaussianBlur(face_image, (99, 99), 30)
    img[top:bottom, left:right] = face_image
  return img

# Blur the face area in the images
blurred_imgElon = blur_face(imgElon.copy())
blurred_imgTest2 = blur_face(imgTest2.copy())
blurred_imgTest3 = blur_face(imgTest3.copy())

# Display the blurred images
cv2_imshow(blurred_imgElon)
cv2_imshow(blurred_imgTest2)
cv2_imshow(blurred_imgTest3)
cv2.waitKey(0)
```

```python
import cv2
import face_recognition

imgElon = face_recognition.load_image_file('ImagesBasic/Elon Musk.jpg')
imgElon = cv2.cvtColor(imgElon,cv2.COLOR_BGR2RGB)
imgTest = face_recognition.load_image_file('ImagesBasic/Bill gates.jpg')
imgTest = cv2.cvtColor(imgTest,cv2.COLOR_BGR2RGB)

faceLoc = face_recognition.face_locations(imgElon)[0]
encodeElon = face_recognition.face_encodings(imgElon)[0]
cv2.rectangle(imgElon,(faceLoc[3],faceLoc[0]),(faceLoc[1],faceLoc[2]),(255,0,255),2)

faceLocTest = face_recognition.face_locations(imgTest)[0]
encodeTest = face_recognition.face_encodings(imgTest)[0]
cv2.rectangle(imgTest,(faceLocTest[3],faceLocTest[0]),(faceLocTest[1],faceLocTest[2]),(255,0,255),2)

results = face_recognition.compare_faces([encodeElon],encodeTest)
faceDis = face_recognition.face_distance([encodeElon],encodeTest)
print(results,faceDis)
cv2.putText(imgTest,f'{results} {round(faceDis[0],2)}',(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,0,255),2)

cv2.imshow('Elon Musk',imgElon)
cv2.imshow('Elon Test',imgTest)
cv2.waitKey(0)
```
